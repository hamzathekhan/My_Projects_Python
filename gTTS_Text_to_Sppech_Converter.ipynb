```python
import gtts
from time import sleep
from IPython.display import Audio
```

```python
questions = [
    "Hey This is Lisa, What is your name?",
    "How old are you",
    "What is your favorite color?",
    "What is your favorite food?",
    "In which class you read?",
    "Where do you live?",
    "Do you love Songs?"
]
```

```python
def speech(question):
  tts = gtts.gTTS(question)
  tts.save("question.mp3")
  return Audio("question.mp3",autoplay=True)
```

```python
def user_speech(user_text):
  tts = gtts.gTTS(user_text)
  tts.save("answer.mp3")
  return Audio("answer.mp3",autoplay=True)
```

```python
for question in questions:
  display(speech(question))
  sleep(2)
  user_text = input("Answer:")
  display(user_speech(user_text))
  sleep(2)
  print("=========================")
```
